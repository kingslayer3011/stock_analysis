```python
# Import Required Libraries
import yfinance as yf
import pandas as pd
```

```python
# Fetch Cash Flow Statement for Apple (AAPL)
ticker = yf.Ticker('AAPL')
cashflow = ticker.cashflow
cashflow = cashflow.copy()  # Ensure we can manipulate it
cashflow.head()
```

```python
# Display All Rows in Cash Flow Statement
print('All available row names (index) in Apple cash flow statement:')
for row in cashflow.index:
    print(row)
```

```python
# Search for Rows Related to 'Appreciation'
search_term = 'appreciation'
matching_rows = [row for row in cashflow.index if search_term.lower() in row.lower()]

if matching_rows:
    print(f"Rows containing '{search_term}':")
    for row in matching_rows:
        print(row)
else:
    print(f"No rows containing '{search_term}' found.")
```
